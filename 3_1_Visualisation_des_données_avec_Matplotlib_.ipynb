D’après la [documentation officielle](https://matplotlib.org/gallery/index.html) :

`matplotlib.pyplot` est une collection de fonctions « style commandes » qui fait fonctionner Matplotlib comme MATLAB. Chaque fonction de pyplot apporte une modification à une figure : par exemple, crée une figure, crée une zone de traçage dans une figure, trace des lignes dans une zone de traçage, décore le graphique avec des labels, etc.

`pyplot` est surtout destiné aux graphiques interactifs et aux cas simples de génération de graphiques par code.

![](https://miro.medium.com/max/2000/1*swPzVFGpYdijWAmbrydCDw.png)

```python
%matplotlib inline
# Ceci est destiné à l’interpréteur IPython afin d’afficher le graphique dans Jupyter.

import pandas as pd
import matplotlib.pyplot as plt
```

### Importe à nouveau le DataFrame, lis‑le dans un DataFrame pandas et assigne‑le à df.

```python
df = pd.read_csv('data/Penguins/penguins_clean.csv')
```

### Rafraîchis ta mémoire sur l’apparence des données

```python
df.head()
```

### Utiliser DataFrame.plot() dans pandas

L’objet DataFrame de pandas possède une méthode `plot()` qui fournit des graphiques de base de différents types, notamment : 'line', 'bar', 'hist', 'box', etc. Tu peux aussi définir des paramètres pour contrôler la mise en page et les labels du graphique.

`plot()` utilise `matplotlib.pyplot` en arrière‑plan, ce qui rend le tracé de données dans un DataFrame beaucoup plus facile.

Cette page te sera très utile :
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html

#### Exemple : box plot en général

```python
df.plot(kind='box')
```

Les échelles de nos données ne s’alignent pas très bien. Pour faciliter le tracé, on va ignorer la masse corporelle des manchots.

```python
df.drop(["body_mass_g"], axis=1).plot(kind='box')
```

#### Meilleure présentation : taille de la figure, ajoute un titre et une légende

```python
df.drop(["body_mass_g"], axis=1).plot(kind='box', figsize=(10,8), title='Box plot of different measurements of species of penguin', legend=True)
```

#### Créer des sous‑graphes

```python
df.plot(kind='box',
        subplots=True, layout=(2,2),
        figsize=(10,8), title='Box plot of different measurements of species of penguin', legend=True)
```

### Divise les données en 3 types en conséquence

```python
df['species'].unique()
```

```python
df_adelie = df[df['species'] == 'Adelie']
```

```python
df_chinstrap = df[df['species'] == 'Chinstrap']
```

```python
df_gentoo = df[df['species'] == 'Gentoo']
```

### Exemple de scatter plot : longueur et largeur du bec

```python
plt.scatter(df_adelie['bill_length_mm'], df_adelie['bill_depth_mm'], c='r')
plt.scatter(df_chinstrap['bill_length_mm'], df_chinstrap['bill_depth_mm'], c='g')
plt.scatter(df_gentoo['bill_length_mm'], df_gentoo['bill_depth_mm'], c='b')
```

#### Meilleure présentation : taille de la figure, ajoute des labels et une légende

```python
plt.figure(figsize=(10,8)) # définir la taille du graphique

plt.scatter(df_adelie['bill_length_mm'], df_adelie['bill_depth_mm'], c='r')
plt.scatter(df_chinstrap['bill_length_mm'], df_chinstrap['bill_depth_mm'], c='g')
plt.scatter(df_gentoo['bill_length_mm'], df_gentoo['bill_depth_mm'], c='b')

ax = plt.gca() #La méthode gca permet au reste du code de faire référence au graphique que nous avons créé

ax.set_xlabel('Bill Length (mm)')
ax.set_ylabel('Bill Width (mm)')
ax.set_title('Bill Length and Width for Different Species of Penguin')

ax.legend(('adelie', 'chinstrap', 'gentoo'))
```

### Exemple d’histogramme : longueur du bec

```python
plt.figure(figsize=(10,8))

plt.hist(df_adelie['bill_length_mm'], color='r', alpha=.5) # alpha règle la transparence du graphique
plt.hist(df_chinstrap['bill_length_mm'], color='g', alpha=.5)
plt.hist(df_gentoo['bill_length_mm'], color='b', alpha=.5)

ax = plt.gca()

ax.set_xlabel('Bill Length (mm)')
ax.set_title('Histogram of Bill Length for Different Species of Penguin')

ax.legend(('adelie', 'chinstrap', 'gentoo'))
```

### Exemple : créer des sous‑graphes

Créer des sous‑graphes uniquement avec `plt` est un peu plus compliqué. C’est considéré comme plus avancé et ça demande de comprendre les « briques » qui composent un graphique. Ne t’en fais pas si tu trouves ça difficile : tu peux toujours suivre l’exemple, puis essayer toi‑même pour mieux comprendre ce qu’il se passe.

L’exemple ci‑dessous trace côte à côte les histogrammes de la longueur et de la largeur du bec.

```python
# D'abord, nous devons décider combien de sous-graphiques nous souhaitons et comment ils sont disposés
# supposons que nous les voulions côte à côte (c’est-à-dire 1 ligne et 2 colonnes)

fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, figsize=(15,8))

# cela va créer un objet figure (qui correspond à toute la zone du graphique)
# et 2 axes (qui sont les 2 sous-graphiques nommés ax0 et ax1)

# Nous pouvons maintenant y placer les graphiques en conséquence

### pour ax0 ###

ax0.hist(df_adelie['bill_length_mm'], color='r', alpha=.5)
ax0.hist(df_chinstrap['bill_length_mm'], color='g', alpha=.5)
ax0.hist(df_gentoo['bill_length_mm'], color='b', alpha=.5)

ax0.set_xlabel('Bill Length (mm)')
ax0.set_title('Histogram of Bill Length for Different Species of Penguin')

ax0.legend(('adelie', 'chinstrap', 'gentoo'))

### pour ax1 ###

ax1.hist(df_adelie['bill_depth_mm'], color='r', alpha=.5)
ax1.hist(df_chinstrap['bill_depth_mm'], color='g', alpha=.5)
ax1.hist(df_gentoo['bill_depth_mm'], color='b', alpha=.5)

ax1.set_xlabel('Bill Width (mm)')
ax1.set_title('Histogram of Bill Width for Different Species of Penguin')

ax1.legend(('adelie', 'chinstrap', 'gentoo'))

plt.show() # après avoir construit ce que nous voulons pour les deux axes, il faut utiliser la méthode show() pour afficher les graphiques
```
