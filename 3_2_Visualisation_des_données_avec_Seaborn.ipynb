## À propos de seaborn
Seaborn est une bibliothèque de visualisation de données en Python basée sur matplotlib. Elle fournit une interface de haut niveau pour tracer des graphiques statistiques attrayants et informatifs, très puissante pour visualiser des données catégorielles.

![](https://d1rwhvwstyk9gu.cloudfront.net/2017/07/seaburn-1.png)

Nous allons utiliser le fichier [Pokemon.csv](https://gist.github.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6). Jetons un œil aux données :

```python
import pandas as pd

pokemon_df = pd.read_csv('data/Pokemon/pokemon.csv', index_col=0)
pokemon_df.head(10)
```

Par exemple, on veut comparer l’attaque (Attack) des différents types de Pokémon pour voir si un type est généralement plus puissant que les autres :

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(x="Type 1", y="Attack", data=pokemon_df);
```

Quand on importe, on abrège généralement 'seaborn' en 'sns'. (C’est une référence à [West Wing / Rob Lowe](https://en.wikipedia.org/wiki/Sam_Seaborn) !) Remarque qu’il faut aussi importer matplotlib.pyplot car Seaborn est une bibliothèque qui s’appuie sur matplotlib. On obtient un graphique mais il est peu lisible, ajoutons quelques réglages pour l’améliorer.

**Essaie : ajoute `aspect=2.5` comme dernier argument dans l’appel à `sns.catplot` ci‑dessous**

```python
sns.catplot(x="Type 1", y="Attack", data=pokemon_df);
```

Tu peux voir qu’en ajoutant 'aspect' on rend le graphique plus large. La largeur est égale à 'aspect * height', donc en augmentant 'aspect' on augmente la largeur. C’est l’un des paramètres de configuration possibles. Pour la liste complète et les détails, consulte la [documentation officielle](https://seaborn.pydata.org/generated/seaborn.catplot.html#seaborn.catplot), mais on va présenter quelques options courantes.

Par exemple, ici on voit qu’il y a un décalage aléatoire sur l’axe des x pour tous les points, afin d’éviter que les points se chevauchent. C’est dû au paramètre 'jitter' qui est à True par défaut. Désactivons‑le pour voir la différence :

**Essaie : ajoute `jitter=False` comme dernier argument dans l’appel à `sns.catplot` ci‑dessous**

```python
sns.catplot(x="Type 1", y="Attack", data=pokemon_df, aspect=2.5);
```

On a maintenant un graphique où les points sont alignés selon leurs catégories, sans décalage sur l’axe des x. Le choix dépend de si la « population » des valeurs (par ex. Attack) est importante. Ici, on veut connaître la distribution d’Attack par Type, donc garder 'jitter' activé peut être utile, ou mieux encore, on peut davantage étaler les points pour montrer la distribution :

**Essaie : ajoute `kind="swarm"` comme dernier argument dans l’appel à `sns.catplot` ci‑dessous**

```python
sns.catplot(x="Type 1", y="Attack", data=pokemon_df, aspect=2.5);
```

On peut le faire en réglant 'kind' sur 'swarm' pour éviter que les points se chevauchent. L’inconvénient est que ce type de tracé demande plus d’espace horizontal. Si on ne veut pas un graphique trop large (par ex. pour un document), on peut le tourner de 90° en inversant x et y, et ajuster aussi aspect et height :

**Essaie : échange `x` et `y`, et ajoute `height=12, aspect=0.6, kind="swarm"` aux arguments du `sns.catplot` ci‑dessous**

```python
sns.catplot(x="Type 1", y="Attack", data=pokemon_df);
```

Voici quelques observations jusqu’ici :

1. Pour certains Types, comme Psychic, l’attaque (Attack) a une plage très large avec une longue « queue » (c.-à-d. que quelques Pokémon de type Psychic ont une très forte attaque alors que la plupart non).

2. À l’inverse, le type Poison se situe majoritairement entre 40 et 110 en Attack.

3. En général, les types Dragon ont plus de puissance d’attaque que les Fées, mais il y a 2 Pokémon de type Fées qui ont une attaque plus élevée.

Allons plus loin : j’ai une théorie selon laquelle les Pokémon légendaires sont plus puissants. Colorons selon 'Legendary' pour voir si le fait qu’un Pokémon soit légendaire influence son Attack :

**Essaie : ajoute `hue="Legendary"` comme dernier argument dans l’appel à `sns.catplot` ci‑dessous**

```python
plt.figure(figsize=(15, 6))
sns.stripplot(x="Type 1", y="Attack", data=pokemon_df, size=7)
```

Ah ha ! On voit que beaucoup des Pokémon de type Psychic qui ont une Attack supérieure aux autres sont en fait des Pokémon légendaires. C’est aussi le cas pour les types Ground et Flying.

Parfois, on veut avoir plusieurs graphiques dans une même figure pour comparer. Une façon de le faire avec seaborn est d’utiliser FacetGrid. La classe FacetGrid est utile pour visualiser la distribution d’une variable ou la relation entre plusieurs variables séparément dans des sous‑ensembles de ton jeu de données. Ci‑dessous, on va utiliser FacetGrid pour voir s’il y a des différences dans notre analyse précédente selon les différentes générations.

Pour créer une FacetGrid, on peut faire ceci :

```python
g = sns.FacetGrid(pokemon_df, col="Generation")
```

On voit qu’on a 6 zones de tracé, ce qui correspond au nombre de générations différentes présentes dans les données
(on peut vérifier quelles sont les différentes générations comme ceci) :

```python
pokemon_df["Generation"].unique()
```

Cependant, on préfère aligner ces graphiques verticalement plutôt qu’horizontalement.

**Essaie : remplace `col` par `row` dans le `sns.FacetGrid` ci‑dessous**

```python
g = sns.FacetGrid(pokemon_df, col="Generation")
```

Ok, maintenant qu’on a la mise en page, comment ajoute‑t‑on le graphique ? Pour certains graphiques, on peut utiliser la méthode [FacetGrid.map()](https://seaborn.pydata.org/generated/seaborn.FacetGrid.map.html#seaborn.FacetGrid.map), par exemple avec sns.countplot pour compter combien de Pokémon il y a dans chaque type :

```python
g = sns.FacetGrid(pokemon_df, row="Generation", aspect=3.5)
g.map(sns.countplot, "Type 1");
```

Mais avec sns.catplot que nous avons utilisé plus tôt, c’est encore plus simple. Comme catplot est déjà une FacetGrid, on peut directement lui ajouter le paramètre `row` ou `col`.

**Essaie : ajoute `row="Generation"` comme dernier argument dans le `sns.catplot` ci‑dessous**

```python
plt.figure(figsize=(15, 6))
sns.stripplot(x="Type 1", y="Attack", data=pokemon_df, size=7, hue="Legendary")
```

Tu vois maintenant que dans chaque génération, les Pokémon légendaires sont des valeurs extrêmes avec des puissances d’attaque supérieures à celles des autres de leur propre génération. Pour plus de détails sur l’utilisation des FacetGrid, consulte la documentation officielle ici : https://seaborn.pydata.org/tutorial/axis_grids.html
