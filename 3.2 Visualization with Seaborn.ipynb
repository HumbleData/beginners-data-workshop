## Seaborn 소개
Seaborn은 matplotlib 위에 구축된 파이썬 데이터 시각화 라이브러리입니다. 보기 좋고 유용한 통계 그래프를 쉽게 그릴 수 있는 고수준 인터페이스를 제공하며, 범주형 데이터를 시각화할 때 특히 강력합니다.

![](https://d1rwhvwstyk9gu.cloudfront.net/2017/07/seaburn-1.png)


이 노트북에서는 [Pokemon.csv](https://gist.github.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6) 데이터를 사용할 것입니다. 먼저 데이터를 살펴보겠습니다:


```python
import pandas as pd

pokemon_df = pd.read_csv('data/Pokemon/pokemon.csv', index_col=0)
pokemon_df.head(10)
```

예를 들어, 포켓몬의 타입별로 Attack(공격)을 비교해 보며 어떤 타입이 전반적으로 더 강한지 알아보겠습니다:


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(x="Type 1", y="Attack", data=pokemon_df);
```

보통 'seaborn'은 간단히 'sns'로 임포트합니다. (드라마 [웨스트 윙의 샘 시본](https://en.wikipedia.org/wiki/Sam_Seaborn)에서 따온 별칭이에요!) Seaborn은 matplotlib 위에 있는 라이브러리이므로 `matplotlib.pyplot`도 함께 임포트해야 합니다. 지금 얻은 플롯은 조금 조악하고 읽기 어려우니, 보기 좋게 몇 가지 설정을 추가해 봅시다.


**시도해 보기: 아래 `sns.catplot` 호출의 마지막 인자로 `aspect=2.5` 를 추가해 보세요**


```python
sns.catplot(x="Type 1", y="Attack", data=pokemon_df);
```

즉, 'aspect'를 추가하면 플롯이 더 넓어집니다. 플롯의 너비는 'aspect * height'로 계산되므로, 'aspect' 값을 키우면 플롯의 가로 폭이 커집니다. 이것은 플롯에 추가할 수 있는 여러 설정 중 하나입니다. 전체 옵션 목록과 상세 설명은 [공식 문서](https://seaborn.pydata.org/generated/seaborn.catplot.html#seaborn.catplot)를 참고하세요. 여기서는 자주 쓰이는 몇 가지만 간단히 소개하겠습니다.


예를 들어, 현재 점들이 서로 겹치지 않도록 모든 점에 무작위 x축 오프셋이 적용되어 있습니다. 이는 기본값이 True인 'jitter' 설정 덕분입니다. 이 옵션을 꺼 보면 어떻게 보이는지 확인해 봅시다:


**시도해 보기: 아래 `sns.catplot` 호출의 마지막 인자로 `jitter=False` 를 추가해 보세요**


```python
sns.catplot(x="Type 1", y="Attack", data=pokemon_df, aspect=2.5);
```

이제 x축 오프셋 없이, 점들이 각 범주(category)에 따라 정렬된 플롯을 얻었습니다. 어느 방식을 사용할지는 값의 분포(예: Attack)가 중요한지에 달려 있습니다. 우리의 경우 각 Type에서 Attack이 어떻게 분포하는지 보고 싶으므로 'jitter'를 켜 두는 것이 좋을 수 있고, 더 나아가 점들을 더 퍼뜨려 분포를 더 잘 보이게 만드는 방법도 있습니다:


**시도해 보기: 아래 `sns.catplot` 호출의 마지막 인자로 `kind="swarm"` 을 추가해 보세요**


```python
sns.catplot(x="Type 1", y="Attack", data=pokemon_df, aspect=2.5);
```

이는 'kind'를 'swarm'으로 설정하면 점들이 서로 겹치지 않게 배치되어 해결할 수 있습니다. 단점은 가로 공간을 더 많이 차지한다는 점입니다. 만약 지면 제한 때문에 플롯을 너무 넓게 만들 수 없다면, x와 y를 뒤바꿔 90도를 회전시키고, aspect와 height를 함께 조정하여 해결할 수 있습니다:


**시도해 보기: 아래 `sns.catplot` 인자에서 `x` 와 `y` 를 바꾸고, `height=12, aspect=0.6, kind="swarm"` 을 추가해 보세요**


```python
sns.catplot(x="Type 1", y="Attack", data=pokemon_df);
```

지금까지 다음과 같은 점들을 관찰할 수 있습니다:

1. 예를 들어 Psychic 타입은 Attack의 범위가 매우 넓고 긴 꼬리(long tail)를 가집니다(일부 Psychic 타입은 공격력이 매우 높지만, 대부분은 그렇지 않음).

2. 반면 Poison 타입은 대체로 Attack 값이 40–110 범위에 분포합니다.

3. 일반적으로 Dragon 타입이 Fairy 타입보다 Attack이 더 높은 경향이 있지만, Attack이 더 높은 Fairy 타입도 2개 있습니다.


하지만 좀 더 깊이 들여다보고 싶습니다: 전설의 포켓몬(Legendary)이 더 강하다는 가설이 있어요. 포켓몬이 Legendary인지 여부에 따라 색을 구분해 보면 Attack과 어떤 관련이 있는지 알 수 있겠죠:


**시도해 보기: 아래 `sns.catplot` 호출의 마지막 인자로 `hue="Legendary"` 를 추가해 보세요**


```python
plt.figure(figsize=(15, 6))
sns.stripplot(x="Type 1", y="Attack", data=pokemon_df, size=7)
```

아하! Attack이 유독 높은 Psychic 타입의 상당수가 사실 Legendary 포켓몬이라는 것을 볼 수 있습니다. 이 현상은 Ground 타입과 Flying 타입에서도 나타납니다.

비교를 위해 하나의 그림 안에 여러 개의 플롯을 배치하고 싶을 때가 있습니다. seaborn에서는 이를 위해 FacetGrid를 사용할 수 있습니다. FacetGrid 클래스는 데이터셋의 하위 집합별로 어떤 변수의 분포를 보거나, 여러 변수 간의 관계를 각각 분리해 시각화할 때 유용합니다. 아래에서는 앞서 수행한 분석이 세대(Generation)에 따라 어떻게 달라지는지 FacetGrid로 살펴보겠습니다.


FacetGrid를 만들기 위해서는 다음과 같이 할 수 있습니다:


```python
g = sns.FacetGrid(pokemon_df, col="Generation")
```

보면 플롯 영역이 6개 생성되었고, 이는 우리가 가진 서로 다른 Generation의 개수와 일치합니다
(서로 다른 Generation 값은 아래처럼 확인할 수 있습니다):


```python
pokemon_df["Generation"].unique()
```

하지만 가로로 나열하는 대신, 플롯을 세로로 정렬하고 싶습니다.


**시도해 보기: 아래 `sns.FacetGrid` 에서 `col` 을 `row` 로 바꿔 보세요**


```python
g = sns.FacetGrid(pokemon_df, col="Generation")
```

좋습니다, 이제 레이아웃은 준비되었습니다. 그럼 플롯은 어떻게 넣을까요? 일부 플롯은 [FacetGrid.map()](https://seaborn.pydata.org/generated/seaborn.FacetGrid.map.html#seaborn.FacetGrid.map) 메서드로 배치할 수 있습니다. 예를 들어, sns.countplot을 이용해 타입별 포켓몬 수를 세어 볼 수 있습니다:


```python
g = sns.FacetGrid(pokemon_df, row="Generation", aspect=3.5)
g.map(sns.countplot, "Type 1");
```

하지만 앞에서 사용했던 sns.catplot을 쓰면 더 간단합니다. catplot 자체가 이미 FacetGrid이므로 `row`나 `col` 설정을 직접 추가할 수 있습니다.


**시도해 보기: 아래 `sns.catplot` 호출의 마지막 인자로 `row="Generation"` 을 추가해 보세요**


```python
plt.figure(figsize=(15, 6))
sns.stripplot(x="Type 1", y="Attack", data=pokemon_df, size=7, hue="Legendary")
```

이제 각 세대 안에서 Legendary 포켓몬들이 다른 포켓몬에 비해 매우 높은 공격력을 가진 이상치(outlier)임을 확인할 수 있습니다. FacetGrid 사용에 대한 자세한 내용은 공식 문서를 참고하세요: https://seaborn.pydata.org/tutorial/axis_grids.html
