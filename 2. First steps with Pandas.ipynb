데이터 사이언티스트들은 데이터 작업을 훨씬 쉽게 해 주는 다양한 파이썬 라이브러리를 사용합니다. 대표적으로 다음과 같아요:

| 패키지 | 설명 |
| -- | -- |
| `NumPy` | 수치 계산용 라이브러리로, C 서브루틴을 호출해 무거운 연산을 처리합니다. 파이썬의 생산성과 C의 계산 성능을 모두 얻을 수 있어요. |
| `SciPy` | 과학 계산, 통계 검정 등 다양한 기능을 제공합니다. |
| `pandas` | 데이터 조작의 스위스 아미나이프. 대부분의 PyData 데모에서 볼 수 있어요! NumPy 위에 구축되어 매우 빠릅니다. |
| `matplotlib` | Matlab에서 영감을 받은 오래됐지만 강력한 데이터 시각화 패키지. |
| `Seaborn` | matplotlib 위에 구축된 더 쉽고 간단한(다소 제한적인) 시각화 패키지. |
| `scikit-learn` | 머신러닝 종합 툴킷! 분류, 회귀, 군집, 차원 축소 등. |
| `nltk` and `spacy` | nltk = 자연어 처리 툴킷; spacy는 더 최신이면서도 사용하기 쉬운 자연어 처리 패키지. |
| `statsmodels` | 통계 검정, 시계열 예측 등. R 사용자에게 익숙한 “모델 공식” 인터페이스를 제공합니다. |
| `requests` and `Beautiful Soup` | `requests` + `Beautiful Soup` = 웹 스크레이퍼를 만들기에 훌륭한 조합. |
| `Jupyter` | 주피터 자체도 패키지입니다. 최신 버전: https://pypi.org/project/jupyter/ 예) `conda install jupyter==1.0.0` 로 업그레이드 |

물론 이 밖에도 정말 많은 라이브러리가 있어요.
