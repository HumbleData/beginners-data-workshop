<div class="alert alert-block alert-warning" style="padding: 0px; padding-left: 20px; padding-top: 5px;"><h2 style="color: #301E40">
Packages d'analyse de données
</h2><br>
</div>

Les Data Scientists utilisent une grande variété de bibliothèques en Python qui rendent le travail avec les données beaucoup plus simple. Ces bibliothèques comprennent principalement :

| Package | Description |
| -- | -- |
| `NumPy` | Calculs numériques — fait tout le « heavy lifting » en s’appuyant sur des sous‑programmes en C. Tu bénéficies donc à la fois de la productivité de Python et de la puissance de calcul de C. Le meilleur des deux mondes ! |
| `SciPy` | Calcul scientifique, tests statistiques et bien plus encore ! |
| `pandas` | Le couteau suisse de la manipulation de données. Tu verras probablement pandas dans toute démo PyData ! pandas est construit au‑dessus de NumPy, donc c’est **rapide**. |
| `matplotlib` | Un package de visualisation de données ancien mais puissant, inspiré de Matlab. |
| `Seaborn` | Un package de visualisation plus récent et facile à utiliser (mais plus limité), construit au‑dessus de matplotlib. |
| `scikit-learn` | Ta boutique « machine learning » tout‑en‑un ! Classification, régression, clustering, réduction de dimension et plus encore. |
| `nltk` et `spacy` | nltk = « natural language processing » toolkit ; spacy est un package plus récent pour le traitement du langage naturel et très facile à utiliser. |
| `statsmodels` | Tests statistiques, prévision de séries temporelles et plus. L’interface « model formula » sera familière aux personnes venant de R. |
| `requests` et `Beautiful Soup` | `requests` + `Beautiful Soup` = super combinaison pour créer des web scrapers. |
| `Jupyter` | Jupyter est aussi un package. Regarde la dernière version sur https://pypi.org/project/jupyter/ et mets à jour par ex. avec `conda install jupyter==1.0.0` |

Bien qu’il en existe d’innombrables autres.

Aujourd’hui, on va surtout se concentrer sur la bibliothèque qui représente 99 % de notre travail : `pandas`. pandas est construit sur la vitesse et la puissance de NumPy.
