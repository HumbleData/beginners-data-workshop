세트(Set)
</h2><br>
</div>


**세트(set)는 중복이 없고, 순서가 없으며, 인덱스가 없는 원소들의 모음입니다.**


```python
important_set = set(['me','myself', 'me', 'I'])
important_set
```

***add*** 메서드를 사용해 세트에 항목을 하나 추가할 수 있습니다...


```python
important_set.add('you')
important_set
```

...또는 ***update*** 메서드를 사용해 여러 항목을 한꺼번에 추가할 수 있습니다.


```python
other_set = {'me', 'all of you', 'other people'}
important_set.update(other_set)
important_set
```

세트 관련 메서드는 [여기](https://docs.python.org/2/library/sets.html)에서 확인할 수 있습니다.  
예를 들어, 두 세트의 교집합을 구할 수 있어요.


```python
set_1 = {3, 6, 9, 12, 15, 18, 21, 24, 27, 30}
set_2 = {5, 10, 15, 20, 25, 30}
set_3 = set_1.intersection(set_2)
set_3
```

<div class="alert alert-block alert-warning" style="padding: 0px; padding-left: 20px; padding-top: 5px;"><h2 style="color: #301E40">
None 체크에 "and" 사용하기
</h2><br>
</div>


가끔 결측값(None 또는 np.nan) 때문에 조건식에서 에러가 날 수 있습니다. 예를 들어 age가 비어 있는데 age > 18을 검사하는 경우죠. 이때 요령이 하나 있어요. `and` 연산에서는 첫 번째 항이 False이면 두 번째 항은 평가하지 않습니다(어차피 결과가 False이기 때문). 따라서 첫 번째 항에서 값이 유효한지 먼저 확인하면 에러를 피할 수 있습니다. 예: 


```python
age = None
age > 18 # you will get an error
```

```python
age = None
(age is not None) and (age > 18) # no error
```

이렇게 하면 코드가 더 단순해지고 `if` 구조의 들여쓰기가 줄어듭니다.


```python
if age is not None:
    if age > 18:
        print("have beer")

# 아래 방식이 더 간결합니다

if (age is not None) and (age > 18):
    print("have beer")
```

**함수는 키워드 "def" 다음에 함수 이름을 쓰고, 괄호 안에 매개변수(parameter)를 넣어 정의합니다.**  

리스트 **list_greeting**을 사용해서, 어떤 문자열이 인사말인지 여부를 판단하는 **is_greeting** 함수를 정의해 봅시다.


```python
list_greeting = ['Hallo', 'Bonjour', 'Ola', 'Hello', 'Ciao', 'Ave']

def is_greeting(s):
    """Returns True if s is in list_greeting, else False."""
    if s in list_greeting:
        return True
    else:
        return False 
```

이제 **Ola**와 **Yo**가 인사말인지 확인해 봅시다.


```python
is_greeting('Ola')
```

```python
is_greeting('Yo')
```
