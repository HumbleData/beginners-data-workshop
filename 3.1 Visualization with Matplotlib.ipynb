[Ά│╡ΉΜζ Έυ╕ΉΕε](https://matplotlib.org/gallery/index.html)ΉΩΡ ΈΦ░Έξ┤Έσ┤:

`matplotlib.pyplot`ΉζΑ MatplotlibΉζΕ MATLABΉ▓αΈθ╝ ΉΓυΉγσΊΧι ΉΙα ΉηΙΆ▓Ν ΊΧ┤ Ήμ╝ΈΛΦ ΈςΖΈι╣ΊαΧ ΊΧρΉΙαΈΥνΉζα ΈςρΉζΝΉηΖΈΜΙΈΜν. Ά░Β pyplot ΊΧρΉΙαΈΛΦ Ά╖╕Έο╝(figure)ΉΩΡ ΉΨ┤ΈΨν Έ│ΑΊβΦΈξ╝ Ήζ╝Ήε╝ΊΓ╡ΈΜΙΈΜν. ΉαΙΈξ╝ ΈΥνΉΨ┤, Ά╖╕Έο╝ΉζΕ ΉΔζΉΕ▒ΊΧαΆ│ι, Ά╖╕Έο╝ ΉΧΙΉΩΡ ΊΦΝΈκεΊΝΖ ΉαΒΉΩφΉζΕ ΈπΝΈΥνΆ│ι, Ά╖╕ ΉΧΙΉΩΡ ΉΕιΉζΕ Ά╖╕ΈουΈσ░, ΈιΙΉζ┤Έ╕Φ ΈΥ▒Ήε╝Έκε ΊΦΝΈκψΉζΕ Ά╛╕Έ░ΚΈΜΙΈΜν.

`pyplot`ΉζΑ Ήμ╝Έκε ΈΝΑΊβΦΊαΧ ΊΦΝΈκψΆ│╝ Ά░ΕΈΜρΊΧε ΊΦΕΈκεΆ╖╕ΈηρΉιΒ ΊΦΝΈκψ ΉΔζΉΕ▒ΉΩΡ ΉιΒΊΧσΊΧσΈΜΙΈΜν.


![](https://miro.medium.com/max/2000/1*swPzVFGpYdijWAmbrydCDw.png)


```python
%matplotlib inline
# this is for ipython interpreter to show the plot in Jupyter

import pandas as pd
import matplotlib.pyplot as plt
```

### ΈΞ░Ήζ┤ΊΕ░ΊΦΕΈιΙΉηΕΉζΕ ΈΜνΉΜε Έ╢ΙΈθυΉβΑ pandas DataFrameΉε╝Έκε Ήζ╜Ά│ι dfΉΩΡ ΊΧιΈΜ╣ΊΧαΉΕ╕ΉγΦ.


```python
df = pd.read_csv('data/Penguins/penguins_clean.csv')
```

### ΈΞ░Ήζ┤ΊΕ░Ά░Α ΉΨ┤ΈΨ╗Ά▓Ν ΉΔζΆ▓╝ΈΛΦΉπΑ ΈΜνΉΜε ΊβΧΉζ╕ΊΧαΆ╕░


```python
df.head()
```

### pandasΉΩΡΉΕε DataFrame.plot() ΉΓυΉγσΊΧαΆ╕░

pandasΉζα DataFrame Ά░ζΉ▓┤ΉΩΡΈΛΦ `plot()` ΈσΦΉΕεΈΥεΆ░Α ΉηΙΉΨ┤ ΈΜνΉΨΣΊΧε Ά╕░Έ│╕ Ά╖╕ΈηαΊΦΕΈξ╝ Ά╖╕Έο┤ ΉΙα ΉηΙΉΛ╡ΈΜΙΈΜν. ΉαΙ: 'line', 'bar', 'hist', 'box' ΈΥ▒. ΈαΡΊΧε ΈπνΆ░εΈ│ΑΉΙαΈξ╝ ΉΕνΉιΧΊΧαΉΩυ ΈιΙΉζ┤ΉΧΕΉδΔΆ│╝ ΈιΙΉζ┤Έ╕ΦΉζΕ ΉιεΉΨ┤ΊΧι ΉΙα ΉηΙΉΛ╡ΈΜΙΈΜν.

`plot()`ΉζΑ ΈΓ┤Έ╢ΑΉιΒΉε╝Έκε `matplotlib.pyplot`ΉζΕ ΉΓυΉγσΊΧαΈψΑΈκε DataFrameΉζα ΈΞ░Ήζ┤ΊΕ░Έξ╝ ΊδρΉΦυ ΉΚ╜Ά▓Ν ΉΜεΆ░ΒΊβΦΊΧι ΉΙα ΉηΙΉΛ╡ΈΜΙΈΜν.

ΈΜνΉζΝ Έυ╕ΉΕεΆ░Α ΊΒ░ ΈΠΕΉδΑΉζ┤ ΈΡσΈΜΙΈΜν:
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html


#### ΉαΙΉΜε: Ά╕░Έ│╕ Έ░ΧΉΛν ΊΦΝΈκψ


```python
df.plot(kind='box')
```

ΈΞ░Ήζ┤ΊΕ░Ήζα ΉΛνΉ╝ΑΉζ╝Ήζ┤ ΉΕεΈκε Ήηα ΈπηΉπΑ ΉΧΛΉΛ╡ΈΜΙΈΜν. ΉΜεΆ░ΒΊβΦΈξ╝ ΉεΕΊΧ┤ ΊΟφΆ╖ΕΉζα Ή▓┤ΉνΣ(body mass)ΉζΑ Ήζ╝ΈΜρ ΉιεΉβ╕ΊΧαΆ▓ιΉΛ╡ΈΜΙΈΜν.


```python
df.drop(["body_mass_g"], axis=1).plot(kind='box')
```

#### ΈΞΦ ΈΓαΉζΑ ΊΣεΊαΕ: Ά╖╕Έο╝ ΊΒυΆ╕░, ΉιεΈςσΆ│╝ Έ▓ΦΈκΑ Ή╢ΦΆ░Α


```python
df.drop(["body_mass_g"], axis=1).plot(kind='box', figsize=(10,8), title='Box plot of different measurements of species of penguin', legend=True)
```

#### ΉΕεΈ╕ΝΊΦΝΈκψ ΈπΝΈΥνΆ╕░


```python
df.plot(kind='box',
        subplots=True, layout=(2,2),
        figsize=(10,8), title='Box plot of different measurements of species of penguin', legend=True)
```

ΊΟφΆ╖ΕΉζα ΉΕεΈκε ΈΜνΈξ╕ ΉλΖΈ│ΕΈκε Έ╢ΑΈου Ά╕╕Ήζ┤Έξ╝ Έ░ΧΉΛν ΊΦΝΈκψΉε╝Έκε Έ╣ΕΆ╡ΡΊΧ┤ Έ┤ΖΉΜεΈΜν. Ήζ┤Έξ╝ ΉεΕΊΧ┤ ΉλΖΈ│ΕΈκε Έ╣ΕΆ╡ΡΊΧι ΉΙα ΉηΙΈΠΕΈκζ DataFrameΉζα ΊαΧΊΔεΈξ╝ ΉΧ╜Ά░Ε Έ░ΦΆ┐ΦΉΧ╝ ΊΧσΈΜΙΈΜν. Ά░Β ΉΩ┤Ήζ┤ ΉΕεΈκε ΈΜνΈξ╕ ΉλΖΉζα Έ╢ΑΈου Ά╕╕Ήζ┤Ά░Α ΈΡαΈΠΕΈκζ ΈΞ░Ήζ┤ΊΕ░Έξ╝ ΊΦ╝Έ▓ΩΊΧαΆ▓ιΉΛ╡ΈΜΙΈΜν.


#### ΈΞ░Ήζ┤ΊΕ░ΉΖΜ ΉνΑΈ╣Ε


```python
df_pivot = df.pivot(columns='species', values='bill_length_mm')
# tell the pivot() method to make the 'species' as columns, and using the 'bill_length_mm' as the value
```

```python
df_pivot.sample(10)
```

#### df_pivotΉζα Έ░ΧΉΛν ΊΦΝΈκψ

Ήζ┤Ήιε `df_pivot`ΉΩΡ ΈΝΑΊΧ┤ `plot()`ΉζΕ ΉΓυΉγσΊΧι ΉΙα ΉηΙΉΛ╡ΈΜΙΈΜν. Έ░ΧΉΛν ΊΦΝΈκψΉζΕ ΈπΝΈΥνΈινΈσ┤ ΈπνΆ░εΈ│ΑΉΙα `kind`Έξ╝ 'box'Έκε ΉΕνΉιΧΊΧαΉΕ╕ΉγΦ. ΈαΡΊΧε Έ│┤Ά╕░ ΉλΜΆ▓Ν Έ│┤Ήζ┤ΈΠΕΈκζ ΉιΒΉιΙΊΧε `figsize`ΉβΑ ΉλΜΉζΑ `title`ΉζΕ ΉΕνΉιΧΊΧαΆ│ι, `legend`ΈΠΕ ΉηΛΉπΑ ΈπΙΉΕ╕ΉγΦ.


#### Ή╢ΦΆ░Α ΉΩ░ΉΛ╡

ΈΜνΈξ╕ Ή▓βΈΠΕ(ΉαΙ: Ή▓┤ΉνΣ)Έξ╝ Ά╕░ΉνΑΉε╝Έκε ΊΦ╝Έ▓ΩΊΧαΉΩυ ΉΩυΈθυΈ╢ΕΈπΝΉζα `df_pivot`ΉζΕ ΈπΝΈΥνΉΨ┤ Έ│┤ΉΕ╕ΉγΦ. Έ░ΧΉΛνΊΦΝΈκψ ΈΝΑΉΜι ΊηΙΉΛνΊΗιΆ╖╕Έηρ(hist)ΉζΕ ΉΓυΉγσΊΧαΈΛΦ Ά▓ΔΈΠΕ ΉΜεΈΠΕΊΧ┤ Έ│┤ΉΕ╕ΉγΦ. ΈαΡΊΧε ΉΕεΈ╕ΝΊΦΝΈκψ 3Ά░εΈκε Ά╡υΉΕ▒ΈΡε ΊΦΝΈκψΉζΕ ΈπΝΈΥνΉΨ┤ Ά░Β ΉΕεΈ╕ΝΊΦΝΈκψΉΩΡ ΉΕεΈκε ΈΜνΈξ╕ ΊΟφΆ╖Ε ΉλΖΉζα ΊηΙΉΛνΊΗιΆ╖╕ΈηρΉζΕ Ά╖╕Έιν Έ│┤ΉΕ╕ΉγΦ.


ΉπΑΆ╕ΙΆ╣ΝΉπΑΈΛΦ `matplotlib.pyplot`ΉζΕ ΉπΒΉιΣ ΉΓυΉγσΊΧαΉπΑ ΉΧΛΉΧαΉΛ╡ΈΜΙΈΜν. `df.plot()`ΉζΕ ΉΥ░Έσ┤ ΈπνΉγ░ ΊΟ╕ΈουΊΧαΉπΑΈπΝ, ΈΧΝΈκεΈΛΦ ΊΦΝΈκψΉζΕ ΈΞΦ ΉΕ╕Έ░ΑΊΧαΆ▓Ν ΉιεΉΨ┤ΊΧαΆ▒░ΈΓα ΈΞΦ Έ│╡ΉηκΊΧε Ά╖╕ΈηαΊΦΕΈξ╝ ΈπΝΈΥνΆ│ι ΉΜ╢ΉζΕ ΈΧΝΆ░Α ΉηΙΉΛ╡ΈΜΙΈΜν. ΈΜνΉζΝ ΉΕ╣ΉΖαΉΩΡΉΕεΈΛΦ `matplotlib.pyplot`(Ήζ┤Έψ╕ `plt`Έκε ΉηΕΊΠυΊΛ╕ΈΡρ)ΉζΕ ΉπΒΉιΣ ΉΓυΉγσΊΧ┤ Έ│┤Ά▓ιΉΛ╡ΈΜΙΈΜν.


### ΈΞ░Ήζ┤ΊΕ░ΉΖΜΉζΕ 3Ά░ΑΉπΑ ΉειΊαΧΉε╝Έκε ΈΓαΈΙΕΆ╕░


```python
df['species'].unique()
```

```python
df_adelie = df[df['species'] == 'Adelie']
```

```python
df_chinstrap = df[df['species'] == 'Chinstrap']
```

```python
df_gentoo = df[df['species'] == 'Gentoo']
```

### ΉΓ░ΉιΡΈΠΕ ΉαΙΉΜε: Έ╢ΑΈου Ά╕╕Ήζ┤ΉβΑ ΈΕΙΈ╣Ε ΊΦΝΈκψ


```python
plt.scatter(df_adelie['bill_length_mm'], df_adelie['bill_depth_mm'], c='r')
plt.scatter(df_chinstrap['bill_length_mm'], df_chinstrap['bill_depth_mm'], c='g')
plt.scatter(df_gentoo['bill_length_mm'], df_gentoo['bill_depth_mm'], c='b')
```

#### ΈΞΦ ΈΓαΉζΑ ΊΣεΊαΕ: Ά╖╕Έο╝ ΊΒυΆ╕░, ΈιΙΉζ┤Έ╕ΦΆ│╝ Έ▓ΦΈκΑ Ή╢ΦΆ░Α


```python
plt.figure(figsize=(10,8)) # set the size of the plot

plt.scatter(df_adelie['bill_length_mm'], df_adelie['bill_depth_mm'], c='r')
plt.scatter(df_chinstrap['bill_length_mm'], df_chinstrap['bill_depth_mm'], c='g')
plt.scatter(df_gentoo['bill_length_mm'], df_gentoo['bill_depth_mm'], c='b')

ax = plt.gca() #gca method tell the rest of the code to reference the plot we made

ax.set_xlabel('Bill Length (mm)')
ax.set_ylabel('Bill Width (mm)')
ax.set_title('Bill Length and Width for Different Species of Penguin')

ax.legend(('adelie', 'chinstrap', 'gentoo'))
```

### ΉΩ░ΉΛ╡: ΉΓ░ΉιΡΈΠΕ έΑΦ ΉπΑΈΛΡΈθυΈψ╕ Ά╕╕Ήζ┤ΉβΑ Ές╕Έυ┤Ά▓ΝΈξ╝ ΊΦΝΈκψΊΧαΆ╕░

Ήζ┤Ήιε ΉΩυΈθυΈ╢Ε Ή░ρΈκΑΉηΖΈΜΙΈΜν. ΉπΒΉιΣ ΊΦΝΈκψΉζΕ ΈπΝΈΥνΉΨ┤ Έ│┤ΉΕ╕ΉγΦ. Ή╢Χ ΈιΙΉζ┤Έ╕ΦΆ│╝ Έ▓ΦΈκΑΈΠΕ Ά╝φ ΉΕνΉιΧΊΧαΉΕ╕ΉγΦ.


### ΊηΙΉΛνΊΗιΆ╖╕Έηρ ΉαΙΉΜε: Έ╢ΑΈου Ά╕╕Ήζ┤ ΊΦΝΈκψ


```python
plt.figure(figsize=(10,8))

plt.hist(df_adelie['bill_length_mm'], color='r', alpha=.5) # alpha set the transparency of the plot
plt.hist(df_chinstrap['bill_length_mm'], color='g', alpha=.5)
plt.hist(df_gentoo['bill_length_mm'], color='b', alpha=.5)

ax = plt.gca()

ax.set_xlabel('Bill Length (mm)')
ax.set_title('Histogram of Bill Length for Different Species of Penguin')

ax.legend(('adelie', 'chinstrap', 'gentoo'))
```

### ΉΩ░ΉΛ╡: ΊηΙΉΛνΊΗιΆ╖╕Έηρ έΑΦ Ές╕Έυ┤Ά▓ΝΈξ╝ ΊΦΝΈκψΊΧαΆ╕░

Ήζ┤Ήιε ΉΩυΈθυΈ╢Ε Ή░ρΈκΑΉηΖΈΜΙΈΜν. ΉπΒΉιΣ ΊΦΝΈκψΉζΕ ΈπΝΈΥνΉΨ┤ Έ│┤ΉΕ╕ΉγΦ. ΉιΒΉιΙΊΧε ΊΙυΈςΖΈΠΕ(alpha)Έξ╝ ΉΕνΉιΧΊΧαΆ│ι, ΉαυΈ░ΦΈξ╕ ΈιΙΉζ┤Έ╕ΦΆ│╝ Έ▓ΦΈκΑΈξ╝ ΊΠυΊΧρΊΨΙΈΛΦΉπΑ ΊβΧΉζ╕ΊΧαΉΕ╕ΉγΦ.


### ΉΕεΈ╕ΝΊΦΝΈκψ ΈπΝΈΥνΆ╕░ ΉαΙΉΜε

ΉανΉπΒ `plt`ΈπΝΉε╝Έκε ΉΕεΈ╕ΝΊΦΝΈκψΉζΕ ΈπΝΈΥεΈΛΦ Ά▓ΔΉζΑ Ήκ░Ά╕Ι ΈΞΦ Έ│╡ΉηκΊΧσΈΜΙΈΜν. ΈΜνΉΗΝ Ά│ιΆ╕Κ Ήμ╝ΉιεΉζ┤Έσ░, ΊΦΝΈκψΉζΕ Ά╡υΉΕ▒ΊΧαΈΛΦ Έ╣ΝΈΦσ Έ╕ΦΈκζΉΩΡ ΈΝΑΊΧε Ήζ┤ΊΧ┤Ά░Α ΊΧΕΉγΦΊΧσΈΜΙΈΜν. ΉΨ┤Έι╡Ά▓Ν ΈΛΡΆ╗┤Ήι╕ΈΠΕ Ά┤εΉ░χΉΧΕΉγΦ. ΉΧΕΈηα ΉαΙΉΜεΈξ╝ ΈΦ░Έζ╝ ΊΧ┤ Έ│┤Ά│ι ΉπΒΉιΣ ΉΜεΈΠΕΊΧ┤ Έ│┤Έσ┤ΉΕε Έυ┤ΉΩΘΉζ┤ ΉΨ┤ΈΨ╗Ά▓Ν ΈΠβΉηΣΊΧαΈΛΦΉπΑ Ήζ┤ΊΧ┤Έξ╝ ΈΕΥΊαΑ Έ│┤ΉΕ╕ΉγΦ.

ΉΧΕΈηα ΉαΙΉΜεΈΛΦ Έ╢ΑΈου Ά╕╕Ήζ┤ΉβΑ Έ╢ΑΈου ΈΕΙΈ╣ΕΉζα ΊηΙΉΛνΊΗιΆ╖╕ΈηρΉζΕ ΉλΝΉγ░Έκε ΈΓαΈηΑΊηΙ Έ░░Ή╣αΊΧ┤ Ά╖╕Έο╜ΈΜΙΈΜν


```python
# First, we have to decide how many subplots we want and how they are orientated
# say we want them side by side (i.e. 1 row 2 columns)

fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, figsize=(15,8))

# this will create a figure object (which is the whole plot area)
# and 2 axes (which are the 2 subplots labeled ax0 and ax1)

# Now we can put plots in them accordingly

### for ax0 ###

ax0.hist(df_adelie['bill_length_mm'], color='r', alpha=.5) 
ax0.hist(df_chinstrap['bill_length_mm'], color='g', alpha=.5)
ax0.hist(df_gentoo['bill_length_mm'], color='b', alpha=.5)

ax0.set_xlabel('Bill Length (mm)')
ax0.set_title('Histogram of Bill Length for Different Species of Penguin')

ax0.legend(('adelie', 'chinstrap', 'gentoo'))

### for ax1 ###

ax1.hist(df_adelie['bill_depth_mm'], color='r', alpha=.5) 
ax1.hist(df_chinstrap['bill_depth_mm'], color='g', alpha=.5)
ax1.hist(df_gentoo['bill_depth_mm'], color='b', alpha=.5)

ax1.set_xlabel('Bill Width (mm)')
ax1.set_title('Histogram of Bill Width for Different Species of Penguin')

ax1.legend(('adelie', 'chinstrap', 'gentoo'))

plt.show() # after building what we want for both axes, use show() method to show plots
```
