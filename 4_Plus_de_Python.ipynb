**Un ensemble est une collection d’éléments uniques, non ordonnés et non indexés.**

...ou plusieurs éléments avec la méthode ***update***.

<div class="alert alert-block alert-warning" style="padding: 0px; padding-left: 20px; padding-top: 5px;"><h2 style="color: #301E40">
Utiliser "and" pour vérifier None
</h2><br>
</div>

Parfois, on a des données manquantes (qui sont None ou np.nan) et cela peut provoquer une erreur dans une condition de vérification (par ex. age > 18 alors que age est manquant ou NaN). Il y a cependant une astuce : dans une opération `and`, le deuxième argument n’est pas évalué si le premier est False (le résultat sera de toute façon False). Donc, en vérifiant d’abord si le premier argument est valide, on peut éviter l’erreur. Par exemple :

```python
age = None
age > 18 # tu obtiendras une erreur
```

```python
age = None
(age is not None) and (age > 18) # pas d’erreur
```

L’avantage de faire ainsi est d’avoir un code plus simple et une structure `if` moins indentée

```python
if age is not None:
    if age > 18:
        print("have beer")

# ce qui n’égale pas
if (age is not None) and (age > 18):
    print("have beer")
```

**On peut définir nos propres fonctions avec le mot‑clé "def" suivi du nom de la fonction et de parenthèses contenant le(s) paramètre(s).**  

En utilisant la liste **list_greeting**, on peut définir une fonction **is_greeting** qui décidera si une chaîne est une salutation ou non.

```python
list_greeting = ['Hallo', 'Bonjour', 'Ola', 'Hello', 'Ciao', 'Ave']

def is_greeting(s):
    """Returns True if s is in list_greeting, else False."""
    if s in list_greeting:
        return True
    else:
        return False
```

On peut maintenant vérifier si **Ola** et **Yo** sont des salutations.

```python
is_greeting('Ola')
```

```python
is_greeting('Yo')
```
